```python
import pandas as pd
```

```python
df = pd.read_json("amzn.json").transpose()
df['date'] = df.index
df.columns = ['open', 'high', 'low', 'close', 'volume', 'date']
df = df.reset_index(drop=True)

df['y'] = df["close"] - df["open"]

date_column = df['date'].astype(str)
for index, date in enumerate(date_column):
    date_column[index] = date_column[index].replace("-", " - ")
df['date'] = date_column
df.head(10)
```

```python
df = df.reset_index(drop=True)
trim_index = df['date'][df['date'] == '2012 - 09 - 24'].index[0]
print('Trim index', trim_index)
print(len(df))
df = df.iloc[trim_index:]
print(df['date'][df['date'] == '2012 - 09 - 24'].index[0])
df.head(10)
```
