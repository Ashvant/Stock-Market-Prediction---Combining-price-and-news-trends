```python
import re
from nltk.corpus import stopwords
```

```python
def news_to_words(raw_news):
    letters_only = re.sub("[^a-zA-Z]", " ",raw_news)#replace sngle words in begining of sentence with " "
    words = letters_only.lower().split()#change all to lower case & split
    stops = set(stopwords.words("english"))#removing stopwords
    meaningful_words = [w for w in words if not w in stops] 
    return( " ".join( meaningful_words )) 
```

```python
trainDatasenti['clean_news'] = trainDatasenti['news'].apply(lambda x: news_to_words(x))
testDatasenti['clean_news'] = testDatasenti['news'].apply(lambda x: news_to_words(x))
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print (topic_idx)
        print (" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(clf, count_vect.get_feature_names(), no_top_words)
```
